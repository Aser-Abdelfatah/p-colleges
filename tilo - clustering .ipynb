```python
import pandas as pd
import numpy as np
embeddings = {}
with open("./top_1000.txt", 'r') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings[word] = vector
```

```python
from KMeansClusterer import KMeansClusterer
km = KMeansClusterer(embeddings,30)
```

```python
km.evaluate_clustering("si")
km.evaluate_clustering("sf",100000)
```

```python
clusters = km.get_labled_cluster(10)
print( clusters[5]) # contains lots of numbers and symbols
print(clusters[6]) # contains lots of stuff associated with dates and time
```
